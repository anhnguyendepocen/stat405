```python
from symbulate import *
%matplotlib inline
```

## Independent

```python
X, Y = RV(BoxModel([(-1, 0), (-1, 1), (0, 0), (0, 1), (1, 0), (1, 1)],
                   probs=[0.12, 0.28, 0.06, 0.14, 0.12, 0.28]))
xy = (X & Y).sim(10000)
xy.plot(['tile','marginal'])
xy.cov(), xy.corr()
```

```python
z = (X + Y).sim(10000)
z.plot()
z.mean(), z.var()
```

## Uncorrelated, not independent

```python
X, Y = RV(BoxModel([(-1, 0), (-1, 1), (0, 0), (0, 1), (1, 0), (1, 1)],
                   probs=[0.10, 0.30, 0.10, 0.10, 0.10, 0.30]))
xy = (X & Y).sim(10000)
xy.plot(['tile','marginal'])
xy.cov(), xy.corr()
```

## Exer 20.3

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
(U1 & U2).sim(1000).plot()
```

```python
X = U1 + U2
Y = (U1 & U2).apply(max)
```

```python
xy = (X & Y).sim(10000)
xy.plot(['hist'])
```

```python
def event(xy):
    if (xy[0] < 0.7) & (xy[1] < 0.5):
        return True
    
xy.count(event) / 10000
```

```python
xy.plot(['hist', 'marginal'])
```

```python
x = X.sim(10000)
x.plot()
x.count_lt(0.7) / 10000
```

```python
y = Y.sim(10000)
y.plot()
y.count_lt(0.5) / 10000
```

```python
x_given_Y_eq_05 = (X | (abs(Y - 0.5) <  0.01) ).sim(10000)
x_given_Y_eq_05.plot()
x_given_Y_eq_05.count_lt(0.7) / 10000
```

```python
y_given_X_eq_07 = (Y | (abs(X - 0.7) <  0.01) ).sim(10000)
y_given_X_eq_07.plot()
y_given_X_eq_07.count_lt(0.5) / 10000
```

```python
u = (U2 | (abs(U1 - 0.5) < 0.01) ).sim(10000)
u.plot()
u.count_lt(0.2) / 10000
```

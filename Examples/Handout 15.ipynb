## Exercise 15.4

```python
plt.xlim([-1, 21])
plt.title('Binomial(n, 0.4) distributions')

Binomial(n=5, p=0.4).plot()
Binomial(n=10, p=0.4).plot()
Binomial(n=15, p=0.4).plot()
Binomial(n=20, p=0.4).plot()

plt.legend(['n=5', 'n=10', 'n=15', 'n=20']);
```

```python
Binomial(n=5, p=0.4).var(), Binomial(n=10, p=0.4).var(), Binomial(n=15, p=0.4).var(), Binomial(n=20, p=0.4).var()
```

```python
plt.xlim([-1, 11])
plt.title('Binomial(10, p) distributions')

Binomial(n=10, p=0.5).plot()
Binomial(n=10, p=0.7).plot()
Binomial(n=10, p=0.9).plot()
Binomial(n=10, p=0.05).plot()

plt.legend(['p=0.5', 'p=0.7', 'p=0.9', 'p=0.05']);
```

```python
Binomial(n=10, p=0.5).var(), Binomial(n=10, p=0.7).var(), Binomial(n=10, p=0.9).var(), Binomial(n=10, p=0.05).var()
```

## Exer 10.1 - rolling dice

```python
P = BoxModel([1, 2, 3 , 4], size=2)
X = RV(P, sum)
Y = RV(P, max)
```

```python
xy = (X & Y).sim(10000)
xy.tabulate(normalize=True)
```

```python
xy.plot(jitter=True)
```

```python
xy.plot("tile")
```

```python
xy.plot(["tile", "marginal"])
```

```python
x = X.sim(10000)
x.plot()
x.mean()
```

```python
y = Y.sim(10000)
y.plot()
y.mean()
```

```python
y_given_xeq6 = (Y | (X == 6) ).sim(10000)
y_given_xeq6.plot()
y_given_xeq6.tabulate()
```

```python
y_given_xeq6.mean()
```

```python
x_given_yeq4 = (X | (Y == 4) ).sim(10000)
x_given_yeq4.plot()
x_given_yeq4.tabulate()
```

```python
Z = X * Y
z = Z.sim(10000)
z.plot()
z.tabulate(normalize=True)
```

```python
z.mean()
```

## Exer 10.2 - Uniform and transformations

```python
P = Uniform(0, 1) ** 2
U1, U2 = RV(P)
X = RV(P, sum)
Y = RV(P, max)
Z = RV(P, min)
```

```python
(Y & Z).sim(10000).plot(['hist', 'marginal'])
```

```python
(Z | (abs(Y - 0.7) < 0.01) ).sim(10000).plot()
```

```python
y = Y.sim(10000)
y.plot()
y.mean()
```

```python
(X & Y).sim(10000).plot(['hist', 'marginal'])
```

```python
(X | (abs(Y - 0.7) < 0.01) ).sim(10000).plot()
```

```python
u = (U1 | (abs(Y - 0.7) < 0.01) ).sim(10000)
u.plot()
u.mean()
```

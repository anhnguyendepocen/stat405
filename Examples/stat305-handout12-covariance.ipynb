## Exer 12.1

```python
P = BoxModel([1, 2, 3, 4], size=2)
X = RV(P, sum)
Y = RV(P, max)
xy = (X & Y).sim(10000)
xy.plot('tile')
xy.cov()
```

## Exer 12.3

```python
P = BoxModel(list(range(1, 6+1)), size=10)
W = RV(P, sum)
X = RV(P, count_eq(1))
Y = RV(P, count_eq(6))
Z = X + Y
```

```python
wx = (W & X).sim(10000)
wx.plot('tile')
wx.cov()
```

```python
wy = (W & Y).sim(10000)
wy.plot('tile')
wy.cov()
```

```python
wz = (W & Z).sim(10000)
wz.plot('tile')
wz.cov()
```

## Exer 12.4

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
X = U1
Y = -U1 + 2 * U1 * U2

xy = (X & Y).sim(10000)
xy.plot(['hist', 'marginal'])
xy.cov(), xy.corr()
```

```python
z = (X * Y).sim(10000)
z.plot()
z.mean()
```

```python
x_given_y = (X | (abs(Y - 0.2) < 0.01) ).sim(10000)
x_given_y.plot()
x_given_y.mean()
```

```python
def conditional_uniform():
    x = Uniform(0, 1).draw()
    y = Uniform(-x, x).draw()
    return x, y

P =  ProbabilitySpace(conditional_uniform)
X, Y = RV(P)

xy = (X & Y).sim(100000)
xy.plot(['hist', 'marginal'])
xy.cov()
```

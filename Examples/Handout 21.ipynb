## Exer 21.4

```python
def random_rectangle():
    x = Exponential(1).draw()
    y = Uniform(0, x).draw()
    return x, y

P = ProbabilitySpace(random_rectangle)
X, Y = RV(P)

(X & Y).sim(10000).plot('hist')
```

```python
perimeter = (2 * (X + Y)).sim(10000)
perimeter.plot()
perimeter.mean()
```

```python
area = (X * Y).sim(10000)
area.plot()
area.mean()
```

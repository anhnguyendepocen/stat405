```python
from symbulate import *
%matplotlib inline
```

```python
M, R = RV(BivariateNormal(mean1=640, mean2=610, sd1=80, sd2=70, corr=0.7))
```

```python
(M & R).sim(1000).plot()
```

```python
(M & R).sim(10000).plot(['hist', 'marginal'])
```

```python
m = M.sim(10000)
m.plot()
Normal(640, 80).plot()
m.count_gt(700) / 10000
```

```python
s = (M + R).sim(10000)
s.plot()
Normal(640+610, var=80**2 + 70**2 +2*0.7*80*70).plot()
s.mean(), s.sd(), s.count_gt(1500) / 10000
```

```python
m_given_R_eq_700 = (M | (abs(R - 700) < 5) ).sim(10000)
m_given_R_eq_700.plot()
Normal(640+0.7*80*(700-610)/70, 80*sqrt(1-0.7**2)).plot()
m_given_R_eq_700.mean(), m_given_R_eq_700.sd(), m_given_R_eq_700.count_gt(700) / 10000
```

```python
d = (M - R).sim(10000)
d.plot()
Normal(640-610, var=80**2 + 70**2 -2*0.7*80*70).plot()
d.mean(), d.sd(), d.count_gt(0) / 10000
```

```python
X, I = RV(Normal(0, 1) * BoxModel([-1, 1]))
Y = I * X
Y.sim(10000).plot()
Normal(0, 1).plot()
```

```python
(X & Y).sim(10000).plot()
```

```python
(X & Y).sim(10000).cov()
```

```python
(X + Y).sim(10000).plot()
```

```python
(X + Y).sim(10000).count_eq(0) / 10000
```

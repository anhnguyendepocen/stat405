## Exer 17.3

```python
n = 500
N1 = 100
x = 4

def likelihood(N):
    return Hypergeometric(n, N1=N1, N0=N-N1).pdf(x)

Ns = range(8000, 20000)
likelihood = [likelihood(N) for N in Ns]

from matplotlib import pyplot as plt
plt.plot(Ns, likelihood)
plt.ylim(0, 0.25);
```

## Exer 17.5

```python
x = RV(Hypergeometric(n=1290, N1=1500, N0=1000)).sim(100000)
x.plot()
x.count_leq(720) / 100000, Hypergeometric(n=1290, N1=1500, N0=1000).cdf(720)
```

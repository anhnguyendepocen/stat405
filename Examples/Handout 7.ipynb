## Exer 7.2 - coin flips

```python
P = BoxModel([1, 0], size=3)
X = RV(P, sum)
Y = 3 - X
```

```python
x = X.sim(10000)
x.plot()
x.mean()
```

```python
y = Y.sim(10000)
y.plot()
y.mean()
```

```python
P = BoxModel([1, 0], probs=[0.6, 0.4], size=3)
X = RV(P, sum)
x = X.sim(10000)
x.plot()
x.mean()
```

## Exer 7.4 - Uniform and transformation

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
```

```python
u = U1.sim(10000)
u.plot()
Uniform(0, 1).plot()
u.mean()
```

```python
X = 400 + 1200 * U1
x = X.sim(10000)
x.plot()
Uniform(400, 1600).plot()
x.mean()
```

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
X = U1 + U2
x = X.sim(10000)
x.plot()
x.mean()
```

```python
Y = RV(Uniform(0, 1) ** 2, min)
y = Y.sim(10000)
y.plot()
y.mean()
```

```python
U = RV(Uniform(0, 1))
Z = -log(U)
z = Z.sim(10000)
z.plot()
Exponential(1).plot()
z.mean()
```

## Exer 6.10 - Uniform(0, 1) and transformations

```python
U = RV(Uniform(0, 1))
U.sim(100).plot("rug")
```

```python
U.sim(10000).plot()
Uniform(0, 1).plot()
```

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
U1.sim(10000).plot()
```

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
(U1 + U2).sim(10000).plot()
```

```python
P = Uniform(0, 1) ** 2
X = RV(P, sum)
X.sim(10000).plot()
```

```python
P = Uniform(0, 1) ** 2
Y = RV(P, min)
Y.sim(10000).plot()
```

```python
U = RV(Uniform(0, 1))
X = -log(U)
X.sim(10000).plot()
```

```python
U = RV(Uniform(0, 1))
X = -log(1 - U)
X.sim(10000).plot()
```

## Exer 6.11 - dice rolls

```python
P = BoxModel([1, 2, 3, 4], size=2, replace=True)
X = RV(P, sum)
Y = RV(P, max)
xy = (X & Y).sim(10000)
xy.plot(["tile", "marginal"])
xy.tabulate(normalize=True)
```

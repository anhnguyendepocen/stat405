```python
from symbulate import *
%matplotlib inline
```

```python
U = RV(Uniform(0, 1) ** 2)
Y = exp(-0.5 * U[0] ** 2)
(U[1] < Y).sim(100000).tabulate()
```

```python
U = RV(Uniform(0, 1))
X = -log(1 - U)
X.sim(100).plot('rug')
```

```python
X.sim(10000).plot(['hist','rug'])
Exponential(1).plot()
```

```python
Y = -log(U)
(X & Y).sim(100)
```

```python
Y.sim(10000).plot()
Exponential(1).plot()
```

```python
U = RV(Uniform(0, 1))
X = (1 - U) ** (-1 / 3)
X.sim(10000).plot()
Pareto(b=3).plot()
```

```python
Pm = [[0.2, 0.4, 0.4],
      [0.3, 0.5, 0.2],
      [0.6, 0.4, 0]]

pi0 = [0, 0, 1]

states = [5, 6, 7]

P = MarkovChainProbabilitySpace(Pm, pi0, states)

X = RV(P)

X.sim(1).plot(tmax=100)
```

```python
N = 1000
V5 = X[0:N].apply(count_eq(5)) / (N + 1)
V5.sim(1000).plot()
```

```python
V5 = RandomProcess(P)
for n in range(N+1):
    V5[n] = X[0:n].apply(count_eq(5)) / (n + 1)
    
V5.sim(100).plot(tmax=N)
```

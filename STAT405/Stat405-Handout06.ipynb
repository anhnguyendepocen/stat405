```python
from symbulate import *
%matplotlib inline
```

```python
states = [1,2]

transition_matrix = [[0.128, 0.872],
                     [0.663, 0.337]]

initial_dist = [0.432, 0.568]

X = MarkovChain(transition_matrix,
                initial_dist,
                state_labels=states)
```

```python
X.sim(10)
```

```python
X.sim(1).plot()
```

```python
X.sim(10).plot()
```

```python
X[0].sim(10000).plot()
```

```python
X[1].sim(10000).plot()
```

```python
X[2].sim(10000).plot()
```

```python
from ipywidgets import interact, widgets
from matplotlib import pyplot as plt

def plot_Xn_distribution(n):
    X[n].sim(10000).plot()
    plt.xlim(0.9, 2.1)
    plt.ylim(0, 1)
    plt.show()

interact(plot_Xn_distribution, n = widgets.IntSlider(min=0, max=10, step=1, value=0));
```

```python
states = [1,2]

transition_matrix = [[0.128, 0.872],
                     [0.663, 0.337]]

initial_dist = [0.1, 0.9]

X = MarkovChain(transition_matrix,
                initial_dist,
                state_labels=states)

def plot_Xn_distribution(n):
    X[n].sim(10000).plot()
    plt.xlim(0.9, 2.1)
    plt.ylim(0, 1)
    plt.show()

interact(plot_Xn_distribution, n = widgets.IntSlider(min=0, max=10, step=1, value=0));
```

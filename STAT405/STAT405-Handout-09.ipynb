```python
from symbulate import *
%matplotlib inline
```

```python
states = [0, 1, 2, 3, 4, 5]

Pm = [[1, 0, 0, 0, 0, 0],
      [1/5, 0, 4/5, 0, 0, 0],
      [0, 2/5, 0, 3/5, 0, 0],
      [0, 0, 3/5, 0, 2/5, 0],
      [0, 0, 0, 4/5, 0, 1/5],
      [0, 0, 0, 0, 0, 1]]

pi0 = [0, 0, 0, 1, 0, 0]

P = MarkovChainProbabilitySpace(transition_matrix=Pm, initial_dist=pi0, state_labels=states)

X = RV(P)

X.sim(1).plot(tmax=20)
```

```python
X.sim(3).plot(tmax=20)
```

```python
def hitting_time_in_state(sequence, A):
    return next( (i for i, x in enumerate(sequence) if (x in A) ), 999999999)

hitting_time_in_state((1,2,3,2,3,2,0), [0, 5])
```

```python
A = [0, 5]

T = RV(P, lambda x: hitting_time_in_state(x, A))
t = T.sim(10000)
t.plot()
t.mean(), t.sd()
```

```python
X[T].sim(10000).plot()
```

```python
states = [0, 1, 2, 3, 4, 5]

Pm = [[1, 0, 0, 0, 0, 0],
      [1/5, 0, 4/5, 0, 0, 0],
      [0, 2/5, 0, 3/5, 0, 0],
      [0, 0, 3/5, 0, 2/5, 0],
      [0, 0, 0, 4/5, 0, 1/5],
      [0, 0, 0, 0, 0, 1]]

pi0 = Binomial(5, 0.5).pdf(range(6))

P = MarkovChainProbabilitySpace(transition_matrix=Pm, initial_dist=pi0, state_labels=states)

X = RV(P)

X.sim(1).plot(tmax=20)
```

```python
T = RV(P, lambda x: hitting_time_in_state(x, A))
t = T.sim(10000)
t.plot()
t.mean(), t.sd()
```

```python
X[T].sim(10000).plot()
```

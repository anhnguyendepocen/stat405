## Note: The arrival and interarrival times use Python's zero-based indexing. So `T[0]` is the time of the first arrival, `T[1]` is the time of the second arrival, etc.

```python
P = PoissonProcessProbabilitySpace(rate = 2)
N = RV(P)
T = RV(P, arrival_times)
```

```python
T[0].sim(10000).plot()
Exponential(rate = 2).plot()
```

```python
W = RV(P, interarrival_times)
w = (W[0] & W[1]).sim(10000)
w.plot(['hist', 'marginal'])
w.corr()
```

```python
T[9].sim(10000).plot()
Gamma(shape = 10, rate = 2).plot()
```

```python
N[1 / 60].sim(10000).tabulate(normalize=True)
```

```python
(T[0] | (N[1] == 1) ).sim(1000).plot()
Uniform(0, 1).plot()
```

```python
(N[2] | (N[6] == 5) ).sim(1000).plot()
Binomial(5, 2 / 6).plot()
```

```python
(T[0] & T[1] | (N[1] == 2) ).sim(1000).plot(['hist', 'marginal'])
```

```python
P = Uniform(0, 1) ** 2
X = RV(P, min)
Y = RV(P, max)
(X & Y).sim(1000).plot()
```

```python
P = PoissonProcessProbabilitySpace(7) * PoissonProcessProbabilitySpace(4) * PoissonProcessProbabilitySpace(1)
N = RV(P)
Ntotal = RV(P, sum)
outcome = P.draw()
N[0](outcome).plot()
N[1](outcome).plot()
N[2](outcome).plot()
Ntotal(outcome).plot()
```

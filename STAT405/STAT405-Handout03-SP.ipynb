# Exercise 3.2

```python
nmax = 4
P = BoxModel([-1, 1], size=nmax)
Z = RV(P)
X = RandomProcess(P)
X[0] = 0
for n in range(nmax):
    X[n+1] = X[n] + Z[n]
```

```python
X.sim(1).plot(tmax=nmax+1)
```

```python
X.sim(3).plot(tmax=nmax+1)
```

```python
X.sim(100).plot(tmax=nmax+1)
```

```python
X[2].sim(10000).plot()
```

```python
(X[2]  & X[4]).sim(10000).plot(['tile', 'marginal'])
```

```python
X[2,4].sim(10000).plot(['tile', 'marginal'])
```

```python
(X[2]  & X[4] | (X[3] == 3 ) ).sim(10000).plot(['tile', 'marginal'])
```

```python
(X[2]  & X[4] | (X[3] == 1 ) ).sim(10000).plot(['tile', 'marginal'])
```

```python
N = X.apply(count_gt(0))
N.sim(10000).plot()
```

```python
def first_max(xs):
    xmax = max(xs)
    for i, x in enumerate(xs):
        if x == xmax:
            return i
```

```python
M = X.apply(first_max)
M.sim(10000).plot()
```

```python
(M & N).sim(10000).plot(['tile', 'marginal'])
```

```python
nmax = 100
P = BoxModel([-1, 1], size=nmax)
Z = RV(P)
X = RandomProcess(P)
X[0] = 0
for n in range(nmax):
    X[n+1] = X[n] + Z[n]

N = X.apply(count_gt(0))
M = X.apply(first_max)
((N / nmax) & (M / nmax)).sim(1000).plot(['scatter', 'marginal'])
```

# Discrete time, continuous state

```python
nmax = 4
P = Normal(0, 1) ** nmax
Z = RV(P)
X = RandomProcess(P)
X[0] = 0
for n in range(nmax):
    X[n+1] = X[n] + Z[n]

X.sim(1).plot(tmax=nmax+1)
```

```python
X.sim(100).plot(tmax=nmax+1)
```

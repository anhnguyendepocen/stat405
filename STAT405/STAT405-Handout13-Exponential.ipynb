```python
from symbulate import *
%matplotlib inline
```

```python
Exponential(1).plot()
```

```python
T = RV(Exponential(1))
T.sim(100).plot('rug')
```

```python
T = RV(Exponential(1))
T.sim(1000).plot(['rug', 'hist'])
Exponential(1).plot()
```

```python
U = RV(Uniform(0, 1))
T = -log(U)
T.sim(10000).plot()
Exponential(1).plot()
```

```python
V = 2 * T
V.sim(10000).plot()
Exponential(1 / 2).plot()
```

```python
T = RV(Exponential(rate = 1 / 30))
( (T - 30) | (T > 30) ).sim(10000).plot()
Exponential(rate = 1 / 30).plot()
```

```python
L, T = RV(Exponential(rate = 1 / 6) * Exponential(rate = 1 / 9))
(L & T).sim(10000).plot(['marginal', 'hist'])
```

```python
Y = (L & T).apply(min)
Y.sim(10000).plot()
Exponential(rate = 1 / 6 + 1 / 9).plot()
```

```python
( L < T ).sim(10000).tabulate()
```

```python
(Y | ( L < T) ).sim(10000).plot()
Exponential(rate = 1 / 6 + 1 / 9).plot()
```
